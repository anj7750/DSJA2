```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd
from tqdm import tqdm

# 크롤링 함수
def crawl_naver_press_releases():
    # Selenium WebDriver 초기화
    chrome_option = Options()
    chrome_option.add_experimental_option("detach", True)
    driver = webdriver.Chrome(options=chrome_option)

    # 네이버 뉴스 페이지로 이동
    url = "https://search.naver.com/search.naver?where=news&query=%EB%8B%A4.&sm=tab_opt&sort=0&photo=4&field=0&pd=1&ds=2025.02.03&de=2025.02.10&docid=&related=0&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3A1w&is_sug_officeid=0&office_category=0&service_area=0"
    driver.get(url)
    time.sleep(2)

    # 데이터 저장용 딕셔너리
    data = {
        '제목': [],
        '링크': [],
        '작성 시간': [],
        '요약': [],
        '언론사': []
    }

    # 스크롤 반복 (모든 기사 로드)
    for _ in tqdm(range(20), desc="스크롤 진행 중"):  # 스크롤을 최대 20번 반복
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
        time.sleep(1)

    # 기사 크롤링
    articles = driver.find_elements(By.CSS_SELECTOR, 'div.news_area')
    print(f"총 {len(articles)}개의 기사를 찾았습니다.")

    for article in articles:
        try:
            # 제목
            title_element = article.find_element(By.CSS_SELECTOR, 'a.news_tit')
            title = title_element.text
            link = title_element.get_attribute('href')

            # 작성 시간
            time_element = article.find_element(By.CSS_SELECTOR, 'span.info')
            written_time = time_element.text

            # 요약
            summary_element = article.find_element(By.CSS_SELECTOR, 'a.api_txt_lines.dsc_txt_wrap')
            summary = summary_element.text

            # 언론사
            press_element = article.find_element(By.CSS_SELECTOR, 'a.info.press')
            press = press_element.text

            # 데이터 저장
            data['제목'].append(title)
            data['링크'].append(link)
            data['작성 시간'].append(written_time)
            data['요약'].append(summary)
            data['언론사'].append(press)
        except Exception as e:
            print(f"기사 파싱 중 오류 발생: {e}")
            continue

    driver.quit()
    return pd.DataFrame(data)

# 실행
if __name__ == "__main__":
    print("네이버 뉴스 보도자료 크롤링 시작...")
    press_releases_df = crawl_naver_press_releases()

    # 데이터 저장
    output_file = "naver_press_releases.csv"
    press_releases_df.to_csv(output_file, index=False)
    print(f"크롤링 완료! 결과가 '{output_file}'에 저장되었습니다.")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm

# 데이터 로드
file_path = "processed_titles.csv"
data = pd.read_csv(file_path)

# 제목 리스트 추출
titles = data["정리된_제목"].dropna().tolist()

# TF-IDF 벡터화
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(titles)

# 코사인 유사도를 배치로 계산
threshold = 0.7  # 유사도 기준
batch_size = 1000  # 배치 크기
results = []
visited = set()

for start_idx in tqdm(range(0, len(titles), batch_size), desc="Processing batches"):
    end_idx = min(start_idx + batch_size, len(titles))
    batch_sim = cosine_similarity(tfidf_matrix[start_idx:end_idx], tfidf_matrix)
    
    for row_idx in range(batch_sim.shape[0]):
        global_idx = start_idx + row_idx
        if global_idx in visited:
            continue
        
        # 현재 제목과의 유사도 계산
        similar_indices = [
            i for i, score in enumerate(batch_sim[row_idx]) if score >= threshold and i != global_idx
        ]
        visited.update(similar_indices)
        visited.add(global_idx)
        
        # 대표 제목, 유사한 기사 수, 유사한 기사들의 제목 저장
        similar_titles = [titles[i] for i in similar_indices]
        results.append({
            "대표_제목": titles[global_idx],
            "유사한_기사_수": len(similar_titles) + 1,  # 자기 자신 포함
            "유사_기사_제목": "; ".join(similar_titles)
        })

# 결과를 DataFrame으로 변환
results_df = pd.DataFrame(results)

# 유사한 기사 수로 내림차순 정렬
results_df = results_df.sort_values(by="유사한_기사_수", ascending=False)

# 결과 저장
output_file = "similar_titles_summary.csv"
results_df.to_csv(output_file, index=False, encoding="utf-8-sig")

print(f"결과 파일이 저장되었습니다: {output_file}")
```
